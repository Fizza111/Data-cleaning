## Find null values containing columns with isna()

```python
#Check the null values
df.isna()
```

```python
#check the null values
df.isna().sum()
```

```python
#info check
df.info()
```

```python
#Drop the columns which has too much null values
df=df.drop(['Alley','MasVnrType','FireplaceQu','PoolQC','Fence','MiscFeature',], axis= 1)
 
```

```python
#Fill the null values with mean or median
df['LotFrontage']=df['LotFrontage'].interpolate()
df['MasVnrArea']=df['LotFrontage'].interpolate()
```

```python
#Now check that null columns has deleted
df.info()
```

```python
#check the null values sum of first 40 columns To make sure that null values has been removed
df.isnull().sum().head(40)
```

```python
#Drop the rows of specific columns that have null values
df=df.dropna(subset=['BsmtQual', 'BsmtCond', 
'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Electrical', 'GarageType', 
'GarageYrBlt', 'GarageFinish', 'GarageQual', 'GarageCond'],axis=0)
```

```python
df.info()
```

```python
#to check how many columns have integer values
df.describe().head(20)
```

```python
#Again check the shape
df.shape
```

```python
#Drop undesired column of id
df=df.drop('Id',axis=1)
```

```python
#To check that all the null values have been removed
df.isnull().sum().sum()
```

```python
#Check final shape
df.shape
```

```python
#Check the categories of data in column
df['Condition1'].value_counts()
```

```python
#Check the datatype of first 40 columns
df.dtypes.head(40)
```

```python
#convert all object type columns data to numerical data at once

df[df.select_dtypes(['object']).columns]=df.select_dtypes(['object']).apply(lambda x: pd.factorize(x)[0])
print(df)

```
